# Venue Analysis

```python
#Getting the unique venue
delivery['venue'].unique()
```

```python
#Total runs scored in each venue
venue_runs = pd.DataFrame(delivery.groupby('venue')['runs_of_bat'].sum().sort_values())
```

```python
venue_runs
```

```python
#Total matches played in the venue
venue_matches = delivery[['venue','match_no']].drop_duplicates()
no_of_matches = venue_matches.value_counts()
match_per_venue = pd.DataFrame(no_of_matches.groupby('venue').sum().sort_values())
match_per_venue.rename(columns={0:'No. of Matches'},inplace=True)
match_per_venue
```

```python
# Total wickets taken in the venue
venue_wicket= delivery[['venue', 'wicket_type']].dropna()
wicket_per_venues = pd.DataFrame(venue_wicket.groupby('venue').count())
wicket_per_venues.rename(columns={'wicket_type':'No. of Wickets'},inplace=True)
wicket_per_venues
```

```python
# Count of wickets in each type
wicket_type_count = pd.DataFrame(venue_wicket.groupby('wicket_type').count())
wicket_type_count.rename(columns={'venue':'No. of Wickets'},inplace=True)
wicket_type_count
```

```python
# Count of wicket based on venue and wicket type
wicket_venue_and_type = pd.DataFrame(venue_wicket.groupby(['venue','wicket_type'])['wicket_type'].count())
wicket_venue_and_type = wicket_venue_and_type.rename(columns={'wicket_type':'No. of Wickets'})
wicket_venue_and_type
```

```python
#Total extras scored in each venue
venue_extra = delivery[['venue', 'extras']][delivery['extras']!=0]
venue_extra = pd.DataFrame(venue_extra.groupby('venue').sum())
venue_extra.rename(columns =  {'extras':'total_extras'}, inplace = True)
venue_extra
```

```python
#Total wide scored in each venue
venue_wide = delivery[['venue', 'wide']][delivery['wide']!=0]
venue_wide = pd.DataFrame(venue_extra.groupby('venue').sum())
venue_wide.rename(columns =  {'wide':'total_wide'}, inplace = True)
venue_wide
```

```python
#Total legbyes scored in each venue
venue_legbyes = delivery[['venue', 'legbyes']][delivery['legbyes']!=0]
venue_legbyes = pd.DataFrame(venue_extra.groupby('venue').sum())
venue_legbyes.rename(columns =  {'wide':'total_legbyes'}, inplace = True)
venue_legbyes
```

```python
#Total byes scored in each venue
venue_byes = delivery[['venue', 'byes']][delivery['byes']!=0]
venue_byes = pd.DataFrame(venue_extra.groupby('venue').sum())
venue_byes.rename(columns =  {'total_extras':'total_Byes'}, inplace = True)
venue_byes
```

```python
#Total noballs scored in each venue
venue_noballs = delivery[['venue', 'noballs']][delivery['noballs']!=0]
venue_noballs = pd.DataFrame(venue_extra.groupby('venue').sum())
venue_noballs.rename(columns =  {'total_extras':'total_No_balls'}, inplace = True)
venue_noballs
```

```python
#Final Chart
venue_analysis = pd.merge(match_per_venue, venue_runs, on='venue').merge(wicket_per_venues, on = 'venue').merge(venue_extra, on='venue').merge(venue_wide, on='venue').merge(venue_legbyes, on='venue').merge(venue_byes, on='venue').merge(venue_noballs, on='venue')
venue_analysis.reset_index(inplace = True)
venue_analysis
```

```python
Venue_data = venue_analysis.to_csv('Venue_data.csv')
```

# Striker Analysis


```python
# Total balls faced( the extra balls due to wide and no balls are not included)
balls_faced = delivery[(delivery['wide'] == 0) & (delivery['noballs'] == 0)]['striker'].value_counts()
balls_faced
```

```python
#Total runs scored by the batsman ( will not include any extras)
total_runsscored = delivery.groupby('striker')['runs_of_bat'].sum()
total_runsscored = pd.DataFrame(total_runsscored)
total_runsscored.rename(columns={'runs_of_bat':'Total Runs Scored'},inplace=True)
total_runsscored.reset_index(inplace=False)
total_runsscored
```

```python
# Strike rate of each player = total runs scored / balls faced
strike_rate = (total_runsscored['Total Runs Scored']/balls_faced)*100
strike_rate = pd.DataFrame(strike_rate)
strike_rate.rename(columns={0:'Strike Rate'},inplace=True)
strike_rate['Strike Rate'] = strike_rate['Strike Rate'].round(2)
strike_rate.reset_index(inplace=True)
strike_rate
```

```python
#Total matches played by batsman
no_of_matches = pd.DataFrame(delivery.groupby('striker')['match_no'].nunique())
no_of_matches.rename(columns={'match_no':'No. of Matches'},inplace=True)
no_of_matches.reset_index(inplace=False)
no_of_matches
```

```python
df=delivery[['striker', 'match_no']]
df.drop_duplicates(inplace= True)
df.groupby('striker')['match_no'].count()
```

```python
#Total innings played by batsman
df1 = delivery[['striker', 'match_no', 'innings']]
df1.drop_duplicates(inplace= True)
no_of_innings = pd.DataFrame(df1.groupby('striker')[ 'innings'].count())
no_of_innings.rename(columns={'innings':'No. of Innings'},inplace=True)
no_of_innings.reset_index(inplace=True)
no_of_innings
```

```python
#Total no of times the striker got Dismissed
all_batters = pd.DataFrame(delivery['striker'].unique())
all_batters.rename(columns={0:'striker'},inplace=True)

dismissal = pd.DataFrame(delivery.groupby('striker')['player_dismissed'].count())
dismissal.rename(columns={'player_dismissed':'No. of Dismissal'},inplace=True)
dismissal.reset_index(inplace=False)

no_of_dismissal = pd.merge(all_batters, dismissal, on='striker', how='left')
no_of_dismissal.fillna(0, inplace=True)
no_of_dismissal['No. of Dismissal'] = no_of_dismissal['No. of Dismissal'].astype(int)
no_of_dismissal
```

```python
#Average of striker = Total runs scored / No of times they got dismissed
no_of_dismissal.set_index('striker', inplace=True)

Average_of_players = (total_runsscored['Total Runs Scored']/no_of_dismissal['No. of Dismissal']).round(2)
Average_of_players = pd.DataFrame(Average_of_players)
Average_of_players.rename(columns={0:'Average'},inplace=True)
Average_of_players.reset_index(inplace=True)
Average_of_players
```

```python
#Count of Sixes
six = delivery[['striker', 'runs_of_bat']]
six = six[six['runs_of_bat'] == 6]
six.groupby('striker')['runs_of_bat'].count()
no_of_sixes = pd.DataFrame(six.groupby('striker')['runs_of_bat'].count())
no_of_sixes.rename(columns={'runs_of_bat':'No. of Sixes'},inplace=True)
no_of_sixes.reset_index(inplace=True)
no_of_sixes
```

```python
#Count of Fours
four = delivery[['striker', 'runs_of_bat']]
four = four[four['runs_of_bat'] == 4]
four.groupby('striker')['runs_of_bat'].count()
no_of_fours = pd.DataFrame(four.groupby('striker')['runs_of_bat'].count())
no_of_fours.rename(columns={'runs_of_bat':'No. of Fours'},inplace=True)
no_of_fours.reset_index(inplace=True)
no_of_fours
```

```python
#Count of Fifties
df2 = delivery.groupby(['striker','match_no'])['runs_of_bat'].sum()
df2 = pd.DataFrame(df2)
df2.reset_index(inplace=True)
no_of_fifties = pd.DataFrame(df2[(df2['runs_of_bat'] > 50) & (df2['runs_of_bat'] < 100)].groupby('striker')['runs_of_bat'].count())
no_of_fifties.rename(columns={'runs_of_bat':'No. of Fifties'},inplace=True)
no_of_fifties.reset_index(inplace=True)
no_of_fifties
```

```python
#Count of Hundreds
df2 = delivery.groupby(['striker','match_no'])['runs_of_bat'].sum()
df2 = pd.DataFrame(df2)
df2.reset_index(inplace=True)
no_of_hundreds = pd.DataFrame(df2[df2['runs_of_bat'] > 100].groupby('striker')['runs_of_bat'].count())
no_of_hundreds.rename(columns={'runs_of_bat':'No. of Hundreds'},inplace=True)
no_of_hundreds.reset_index(inplace=True)
no_of_hundreds
```

```python
# Team of the Striker
team = pd.DataFrame(delivery[['striker', 'batting_team']].drop_duplicates())
team
```

```python
# Total count of matches striker was not out
Not_out = (no_of_matches['No. of Matches'] - no_of_dismissal['No. of Dismissal'])

Not_out = pd.DataFrame(Not_out)
Not_out.rename(columns={0:'Not Out'},inplace=True)
Not_out.reset_index(inplace=True)
Not_out
```

```python
#Final Chart
Striker_Table = pd.merge(team, no_of_matches, on = 'striker', how = 'left').merge(no_of_innings, on = 'striker', how = 'left').merge(balls_faced, on = 'striker', how = 'left').merge(Not_out, on = 'striker', how = 'left').merge(total_runsscored, on = 'striker', how = 'left').merge(strike_rate, on = 'striker', how = 'left').merge(Average_of_players, on = 'striker', how = 'left').merge(no_of_dismissal, on = 'striker', how = 'left').merge(no_of_sixes, on = 'striker', how = 'left').merge(no_of_fours, on = 'striker', how = 'left').merge(no_of_fifties, on = 'striker', how = 'left').merge(no_of_hundreds, on = 'striker', how = 'left')
Striker_Table.fillna(0, inplace = True)
Striker_Table[[ 'No. of Sixes', 'No. of Fours', 'No. of Fifties', 'No. of Hundreds']] = Striker_Table[[ 'No. of Sixes', 'No. of Fours', 'No. of Fifties', 'No. of Hundreds']].astype(int)
Striker_Table.drop(170, inplace = True)
Striker_Table.rename(columns = {'striker':'Striker','batting_team':'Team', 'count' : 'Balls Faced','No. of Matches': 'Matches', 'No. of Innings':'Innings', 'Total Runs Scored' : 'Runs','No. of Dismissal': 'Dismissals', 'No. of Sixes': 'Sixes', 'No. of Fours' : 'Fours', 'No. of Fifties' : 'Fifties' , 'No. of Hundreds': 'Hundreds' }, inplace = True)
Striker_Table
```

# Bowler Analysis

```python
delivery
```

```python
# Team of the Bowler
team1 = pd.DataFrame(delivery[['bowler', 'bowling_team']].drop_duplicates())
team1
```

```python
# Total balls bowled by the bowler( Extra ball due to wide and no ball are not included)
balls_bowled = pd.DataFrame(delivery[(delivery['wide'] == 0) & (delivery['noballs'] ==0)]['bowler'].value_counts())
balls_bowled.rename(columns = {'count' : 'Balls'}, inplace = True)
balls_bowled.reset_index(inplace=True)
balls_bowled
```

```python
#Total matches played by bowler
B_no_of_matches = pd.DataFrame(delivery.groupby('bowler')['match_no'].nunique())
B_no_of_matches.rename(columns={'match_no':'No. of Matches'},inplace=True)
B_no_of_matches.reset_index(inplace=True)
B_no_of_matches
```

```python
#Total innings played by bowler
B_no_of_innings = pd.DataFrame(delivery[['bowler' , 'match_no', 'innings']].drop_duplicates().groupby('bowler')['innings'].count())
B_no_of_innings.reset_index(inplace=True)
B_no_of_innings.rename(columns={'innings':'No. of Innings'},inplace=True)
B_no_of_innings
```

```python
#Total over bowled by bowler
balls_bowled['No. of Over'] = balls_bowled['Balls'].apply(lambda x: f"{x // 6}.{x % 6}")
total_over = pd.DataFrame(balls_bowled)
total_over.drop('Balls' , axis=1, inplace = True)
total_over.reset_index(inplace=True)
total_over.drop('index' , axis=1, inplace = True)
total_over['No. of Over'] = total_over['No. of Over'].astype(float)
total_over
```

```python
# #Total wickets taken by bowler( Run out and retired hurt are not included)
legal_wicket_type = delivery[(delivery['wicket_type'] != 'runout') & (delivery['wicket_type'] != 'retired hurt') & (delivery['wicket_type'] != 'obstructing the field')]
total_wicket = pd.DataFrame(legal_wicket_type.groupby('bowler')['wicket_type'].count())
total_wicket.rename(columns={'wicket_type':'No. of Wickets'},inplace=True)
total_wicket.reset_index(inplace=True)
total_wicket
```

```python
# Legal runs bowled = byes and legbyes are zero ( Runs got by legbyes and byes are not included)
filtered = delivery[(delivery['legbyes'] == 0) & (delivery['byes'] == 0)]
legal_runs_given = pd.DataFrame(filtered.groupby('bowler')[['runs_of_bat' , 'extras']].sum())
legal_runs_given.reset_index(inplace=True)
legal_runs_given

```

```python
# Total legal runs given by bowler( runs from bat + extras due to wide and no ball)
total_legal_runs = legal_runs_given['runs_of_bat'] + legal_runs_given['extras']
total_legal_runs = pd.DataFrame(total_legal_runs)
total_legal_runs['bowler'] =legal_runs_given['bowler']
total_legal_runs.rename(columns={0:'Runs Given'},inplace=True)

total_legal_runs
```

```python
# Total extras given by bowler
extra_runs_given = pd.DataFrame(delivery.groupby(['bowler', 'match_no'])[['runs_of_bat' , 'extras', 'legbyes','byes','wide', 'noballs']].sum())
extra_runs_given.reset_index(inplace=True)
extra_runs_given
```

```python
# Matchwise analysis =  find the runs and wicket secured by bowler in every match
match_wise_runs = pd.DataFrame(filtered.groupby(['bowler', 'match_no'])[['runs_of_bat','extras']].sum())
match_wise_runs['runs_with_extras'] = match_wise_runs['runs_of_bat'] + match_wise_runs['extras']
match_wise_runs.reset_index(inplace=True)

match_wise_wicket = pd.DataFrame(legal_wicket_type.groupby(['bowler', 'match_no'])['wicket_type'].count())
match_wise_wicket.reset_index(inplace=True)

match_wise_analysis = pd.merge(match_wise_runs, match_wise_wicket, on = ['bowler', 'match_no'])
match_wise_analysis = pd.DataFrame(match_wise_analysis)
match_wise_analysis.rename(columns={'runs_of_bat':'Runs', 'wicket_type':'Wickets'},inplace=True)
match_wise_analysis.drop('match_no', axis = 1, inplace= True)

match_wise_analysis['bowling_figure'] = match_wise_analysis['Wickets'].astype(str) + '-' + match_wise_analysis['runs_with_extras'].astype(str)
match_wise_analysis

```

```python
# Finding the best bowling figure of bowler = high wicket taken and low runs provided
best_bowling_figures = match_wise_analysis.sort_values(
    by=['Wickets', 'runs_with_extras'], ascending=[False, True]).drop_duplicates(subset='bowler', keep='first')


best_bowling_figures.drop(['Runs' , 'extras' ,'runs_with_extras','Wickets'], axis = 1, inplace= True)
best_bowling_figures
```

```python
# Economy of batsman = Total legal runs provided / Total overs bowled
economy = pd.DataFrame(total_legal_runs['bowler'])
economy = pd.merge(economy, total_over, on = 'bowler', how = 'left').merge(total_legal_runs, on = 'bowler', how = 'left')
economy['Economy'] = (economy['Runs Given'] / economy['No. of Over']).round(2)
economy.drop(['Runs Given', 'No. of Over'], axis = 1, inplace = True)
economy
```

```python
# Four wicket haul = count of matches where bowler took 4 wickets
wicket_per_over = pd.DataFrame(legal_wicket_type.groupby(['bowler', 'match_no'])['wicket_type'].count())
wicket_per_over.reset_index(inplace=True)
four_wicket_hall = pd.DataFrame(wicket_per_over[wicket_per_over['wicket_type'] == 4].groupby('bowler')['wicket_type'].count())
four_wicket_hall.reset_index(inplace=True)
four_wicket_hall
```

```python
# Five wicket haul = count of matches where bowler took 5 wickets
five_wicket_hall = pd.DataFrame(wicket_per_over[wicket_per_over['wicket_type'] == 5].groupby('bowler')['wicket_type'].count())
five_wicket_hall
```

```python
#Final Chart
Bowler_table = pd.merge(team1, B_no_of_matches, on = 'bowler', how = 'left').merge(B_no_of_innings, on = 'bowler', how = 'left').merge(balls_bowled, on = 'bowler', how = 'left').merge(total_over, on = 'bowler', how = 'left').merge(total_legal_runs, on = 'bowler', how = 'left').merge(total_wicket, on = 'bowler', how = 'left').merge(best_bowling_figures, on = 'bowler', how = 'left').merge(economy, on = 'bowler', how = 'left').merge(four_wicket_hall, on = 'bowler', how = 'left').merge(five_wicket_hall, on = 'bowler', how = 'left')
Bowler_table.fillna(0, inplace = True)
Bowler_table.rename(columns = {'bowler' : 'Bowler','bowling_team':'Team', 'No. of Matches' : 'Matches', 'No. of Innings' : 'Innings','No. of Over_x' : 'Overs','Runs Given' : 'Runs', 'No. of Wickets' : 'Wickets', 'bowling_figure' : 'Best_bowling_figure', 'wicket_type_x' : 'Four_wicket_hauls','wicket_type_y' : 'Five_wicket_hauls'}, inplace = True)
Bowler_table.drop('No. of Over_y', axis = 1, inplace = True)
Bowler_table[['Wickets', 'Runs', 'Four_wicket_hauls', 'Five_wicket_hauls']] = Bowler_table[['Wickets', 'Runs', 'Four_wicket_hauls', 'Five_wicket_hauls']].astype(int)
Bowler_table
```
